```python
# Step_0 : Install Needed Packages 
import pandas as pd
import numpy as np 
```

```python
# Step_1 : Select Your Country + Sector as in Your Excel Input Excel Sheet
Country_Sector='US_Clean' 
Sector=(Country_Sector+'.xlsx')
```

```python
# Step_2 : Here Is your TOPSIS Algorithm 


# 1 _ ===> The first step is to read the data;import pandas as pd #import pandas toolkit

DA = pd.read_excel(Sector)
DA.fillna(value=0, inplace=True)

DA2=DA.iloc[:,:2]
data=DA.iloc[:,2:]


# ==> The second step is to normalize the data;
# Get column name
n = list(data.columns)
# normalize
for i in n:
    x = np.power(np.sum(np.power(data[i],2)),0.5)
    data[i] = data[i]/x 



# 2 _  ===> The third step is to obtain an optimal column value vector and a worst value vector;

A1 = []
# Get the optimal column value
for i in n:
    Max = np.max(data[i])
    A1.append(Max)    
#  form    
A1 = np.array(A1)



# 3 _  Get the worst column value
A2 = []
for i in n:
    Min = np.min(data[i])
    A2.append(Min)    
#  form    
A2 = np.array(A2)




# 4 _  ===> The fourth step is to calculate the optimal plan and the worst plan;
# Optimal solution
D1 = []
for i in range(1,len(data)+1):
    d1 = np.power(np.sum(np.power(A1 - data[i-1:i],2).values),0.5)
    D1.append(d1)
#  form
D1 = np.array(D1)

# Calculate the worst solution
D2 = []
for i in range(1,len(data)+1):
    d2 = np.power(np.sum(np.power(A2 - data[i-1:i],2).values),0.5)
    D2.append(d2)
#  form
D2 = np.array(D2)





# 5 _  ===> The fifth step is to calculate the proximity of the evaluation object to the optimal solution;


C = D2/(D1 + D2)



# 6 _  ===> Finally, sort it.


data1 = pd.DataFrame(C, index = list(data.index))
# Sort by C value
data1['rank'] = data1.rank(ascending = False)
# Reset column name
data1.columns = ['score', 'rank']


# data1.to_excel('DATAA_TOPSIS.xlsx')


# Merge Table 
TOPSIS=data1.merge(DA, how='inner', left_index=True, right_index=True)
TOPSIS=TOPSIS.iloc[:,1:]
```

```python
# Step_3 : Save Your Topsis Results in a Excel Sheet  
TOPSIS.to_excel('DATAA_TOPSIS_'+Sector)
```
